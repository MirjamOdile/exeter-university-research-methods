### The Business School


```python
college_name = "The Business School"

urls = ["https://business-school.exeter.ac.uk/about/people/economics/",
        "http://business-school.exeter.ac.uk/about/people/finance/",
        "https://business-school.exeter.ac.uk/about/people/management/",
        "https://business-school.exeter.ac.uk/about/people/site/",
        "https://business-school.exeter.ac.uk/about/people/sustainable-futures/",
        "https://business-school.exeter.ac.uk/about/people/pgr/"]

departments = ["Economics", "Finance and Accounting", "Management",
               "SITE (Science, Innovation, Technology, and Entrepreneurship)",
               "Sustainable Futures", "Business School PGR"]

name = []
website = []
role = []
department = []
college = []

for i, url in enumerate(urls):
    page = requests.get(url)
    tree = html.fromstring(page.content)
    tree.make_links_absolute(url)
    for j in range(1,500):
        path_name = '//*[@class="profile_list"]/tr['+str(j)+']/td[1]/a/text()'
        if tree.xpath(path_name) != []:
            name.append(tree.xpath(path_name)[0])
            path_url = '//*[@class="profile_list"]/tr['+str(j)+']/td[1]/a/@href'
            try:
                website.append(tree.xpath(path_url)[0])
            except:
                website.append("") 
            path_role = '//*[@class="profile_list"]/tr['+str(j)+']/td[2]/text()'
            try:
                role.append(tree.xpath(path_role)[0])
            except: 
                role.append("")
            college.append(college_name)
            department.append(departments[i])
        
businessschool = pd.DataFrame(list(zip(college, department, name, role, website)),
                              columns=['college', 'department', 'name', 'role', 'url'])
businessschool
```

# 2) Merge and save the data

```python
staff = pd.concat([businessschool, cemps, humanities, lifesciences, medicine, ssis])
staff
```

```python
staff.college.value_counts()
```

```python
# Create names for matching
staff["namelast"] = staff.name.str.replace("Professor |Dr ", "", regex = True).str.split().apply(lambda x: x[-1])
staff["namefirst"] = staff.name.str.replace("Professor |Dr ", "", regex = True).str.split().apply(lambda x: x[0])
staff["identifier1"] = staff.namelast + ", " + staff.namefirst.str[0]
staff["identifier2"] = staff.namelast + ", " + staff.namefirst
# Rename url to profile
staff = staff.rename(columns={"url": "profile",})
```

```python
# Drop all exact duplicates
staff = staff.drop_duplicates(subset=None, keep='first', inplace=False)
# Drop partial duplicates (name/department) where the role is "missing" in one of the observations
staff = staff.loc[~((staff.duplicated(['name', 'department'], keep=False)) & (staff.role.isnull()))]
# Drop all remaining department duplicates and keep last role
staff = staff.drop_duplicates(subset=['name', 'department'], keep='last', inplace=False)
# Drop partial duplicates (name/college) where the role is "missing" in last of the observations
staff = staff.loc[~((staff.duplicated(['name', 'college'], keep='last')) & (staff.role.isnull()))]
staff = staff.loc[~((staff.duplicated(['name', 'college'], keep='first')) & (staff.role.isnull()))]
# Drop partial duplicates (name/college) where one role is PhD and the othes is new role
staff = staff.loc[~((staff.duplicated(['name', 'college'], keep= False)) & (staff.role.str.match('PhD')))]
# Drop officers/ student recruitment managers / administrators / coordinators
staff = staff.loc[~staff.role.str.match('Web Officer', na=False)]
staff = staff.loc[~((staff.duplicated(['name', 'college'], keep= False)) & (staff.role.str.contains('Officer')))]
staff = staff.loc[~staff.role.str.contains('(?:Experimental|Scientific|Support|Political) Officer', na=False, regex = True)]
staff = staff.loc[~staff.role.str.contains('Student Recruitment', na=False, regex = True)]
staff = staff.loc[~staff.role.str.contains('Administrator|Coordinator|Marketing, Recruitment, Communications & Global Business Partner', na=False, regex = True)]
# Drop all remaining college duplicates and keep last role
staff = staff.drop_duplicates(subset=['name', 'college'], keep='first', inplace=False)
```
